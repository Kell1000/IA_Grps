# **Exercice 4**

```python
import plotly.express as px
```

```python
#a) Graphiques
fig_math = px.histogram(df, x='math score',
                        nbins=10,
                        title='Histogramme - Math Score')
fig_math.show()

fig_reading = px.histogram(df, x='reading score',
                           nbins=10,
                           title='Histogramme - Reading Score')
fig_reading.show()

fig_writing = px.histogram(df, x='writing score',
                           nbins=10,
                           title='Histogramme - Writing Score')
fig_writing.show()
```

```python
#Boîtes à moustaches (Box plots) → Quartiles et valeurs aberrantes

fig_box = px.box(df,
                 y=['math score', 'reading score', 'writing score'],
                 title='Box Plot des Scores')
fig_box.show()
```

```python
#Graphiques en barres → Comparaisons entre matières
means = df[['math score', 'reading score', 'writing score']].mean().reset_index()
means.columns = ['Subject', 'Mean Score']

fig_bar = px.bar(means,
                 x='Subject',
                 y='Mean Score',
                 title='Comparaison des Moyennes')
fig_bar.show()
```

# **EXERCICE 5**

```python
#a
correlation = df[['math score', 'reading score', 'writing score']].corr()

fig_corr = px.imshow(correlation,
                     text_auto=True,
                     title="Heatmap de Corrélation")
fig_corr.show()
```

```python
#b
fig_scatter = px.scatter(df,
                         x='math score',
                         y='reading score',
                         title='Math vs Reading',
                         trendline='ols')
fig_scatter.show()
```

```python
fig_matrix = px.scatter_matrix(df,
              dimensions=['math score', 'reading score', 'writing score']
,
                                title='Scatter Matrix des Scores')
fig_matrix.show()
```

```python
fig_prep = px.box(df,
                  x='test preparation course',
                  y='math score',
                  title='Impact du Test Preparation Course')
fig_prep.show()
```

```python
fig_lunch = px.box(df,
                   x='lunch',
                   y='math score',
                   title='Impact du Lunch')
fig_lunch.show()
```

```python
fig_parent = px.box(df,
                    x='parental level of education',
                    y='math score',
                    title="Impact de l'Éducation Parentale")
fig_parent.update_layout(xaxis_tickangle=45)
fig_parent.show()
```
