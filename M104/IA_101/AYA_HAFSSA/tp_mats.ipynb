```python
def detect_outliers(df, column):
    Q1 = np.percentile(df[column], 25)
    Q3 = np.percentile(df[column], 75)
    Q2 = np.percentile(df[column], 50)
    IQR = Q3 - Q1

    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lim_inf) | (df[column] > lim_sup)][column]

    return {
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "Limite_inf": lim_inf,
        "Limite_sup": lim_sup,
        "Outliers": outliers
    }

math_out = detect_outliers(df, "math score")
reading_out = detect_outliers(df, "reading score")
writing_out = detect_outliers(df, "writing score")


print(math_out)


print(reading_out)


print(writing_out)
```
